```python
''' How a neural network learns (learning rate)'''
import keras
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model= Sequential([Dense(16,input_shape=(1,),activation='relu'),
                  Dense(32,activation='relu'),
                  Dense(2,activation='softmax')])
```

```python
model.compile(Adam(lr=.0001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
#lr is learning rate
```

```python
''' function to train the model SGD (Stochastic Gradient descend )'''
#It assigns weights in such a way that it brings 
#the loss function as close to zero as possible 
# Leraning rate lr= d(loss)/d(weight)
'''PREPROCESSSING DATA'''
from random import randint
from sklearn.preprocessing import MinMaxScaler
train_labels=[]
train_samples=[]
for i in range(1000):
    random_younger=randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    random_older=randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
for i in range(50):
    random_younger=randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    random_older=randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
train_labels=np.array(train_labels)
train_samples=np.array(train_samples)
scaler=MinMaxScaler(feature_range=(0,1))
scaled_train_samples=scaler.fit_transform((train_samples).reshape(-1,1))
```

```python
model.fit(scaled_train_samples,train_labels,batch_size=10,epochs=20,shuffle=True,verbose=2)
```

```python
# loss in a neural network explained
model.compile(Adam(lr=.0001),loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# adding extra null layers to increase the output size to retain quality
# categories of padding
# valid-no padding
#same padding in order to maintain same size in output as in input
'''MODEL BEFORE PADDING'''
import keras
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense,Flatten
from keras.layers.convolutional import *
model_valid=Sequential([
    Dense(16,activation='relu',input_shape=(20,20,3)),
    Conv2D(32,kernel_size=(3,3),activation='relu',padding='valid'),
    Conv2D(64,kernel_size=(5,5),activation='relu',padding='valid'),
    Conv2D(128,kernel_size=(7,7),activation='relu',padding='valid'),
    Flatten(),
    Dense(2,activation='softmax'),
])
model_valid.summary()
```

```python
#MODEL AFTER PADDING
import keras
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense,Flatten
from keras.layers.convolutional import *
model_valid=Sequential([
    Dense(16,activation='relu',input_shape=(20,20,3)),
    Conv2D(32,kernel_size=(3,3),activation='relu',padding='same'),
    Conv2D(64,kernel_size=(5,5),activation='relu',padding='same'),
    Conv2D(128,kernel_size=(7,7),activation='relu',padding='same'),
    Flatten(),
    Dense(2,activation='softmax'),
])
model_valid.summary()
```

```python
# we notice that after padding that output size is same as input size ie 
#(None,20,20,16) while in without padding output was (None,8,8,128) 
```

```python
'''MAX POOLING'''
# It reduces the dimesionality of the images by reducing the pixels
# we define a nxn region as filter and a stride number n which defines
#how many pixels should our filter move while sliding throuhght the image
#the filter slides and finds max value in current selected region 
# This is called max pooling , a final image is formed with all the max values
# USES OF AMX POOLING
# reduces resolution which reduces parameters and computational load 
# Reduce Overfitting
```

```python
#CODE FOR MAXPOOLING
import keras
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense,Flatten
from keras.layers.convolutional import *
from keras.layers.pooling import *
model=Sequential([
    Dense(16,activation='relu',input_shape=(20,20,3)),
    Conv2D(32,kernel_size=(3,3),activation='relu',padding='same'),
    MaxPooling2D(pool_size=(2,2),strides=2,padding='valid'),
    Conv2D(64,kernel_size=(5,5),activation='relu',padding='same'),
    Flatten(),
    Dense(2,activation='softmax'),
])
model.summary()
```

```python
#BACK PROPOGATION to reduce loss is the proccess in which the SGD modifies the
#weights in the previous layer to reduce the loss function
```

```python
# 1.PASS data to model via forward propogation
# 2. Calculate loos on output
# 3. SGD to minimize the loss
#   - By calculating the gradient of the loss function and updating the weights
#   - Gradient is calculated by backpropogation
```

```python
#WEIGHT INITIALIZATION USING XAVIER INITIAIZATION
model=Sequential([
    Dense(16,input_shape=(1,5),activation='relu'),
    Dense(32,activation='relu',kernel_initializer='glorot_uniform'),
    Dense(2,activation='softmax')
])
model.summary()
# glorot unifor is for XAvier initialization
```

```python
'''REGULARISATION'''
# IT reduces overfitting and variance by penalizing for complexity
# loss+x
# L2 regularization is the most common type
```

```python
#REGULARISATION IN KERAS
import keras
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras import regularizers
model=Sequential([
    Dense(16,input_shape=(1,),activation='relu'),
    Dense(32,activation='relu',kernel_regularizer=regularizers.l2(0.01)),
    Dense(2,activation='sigmoid')
])
model.summary()
```

```python
'''BATCH NORMAILIZATION'''
# In a nurealization we either need to normalize or standardize the data
#BATCH NORM USING KERAS
from keras.models import Sequential
from keras.layers import Dense,Activation,BatchNormalization
model=Sequential([
    Dense(16,input_shape=(1,5),activation='relu'),
    Dense(32,activation='relu'),
    BatchNormalization(axis=1),
    Dense(2,activation='softmax')
])
#beta_initializer : initializer for beta weight
#gamma_initializer : initializer for gamma weight
model.summary()
```
